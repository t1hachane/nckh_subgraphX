```python
import pickle
import torch
from torch_geometric.data import Data
import os

def read_ba_shapes(dataset_path):
    # Check if file exists
    if not os.path.exists(dataset_path):
        raise FileNotFoundError(f"Dataset not found at {dataset_path}")
    
    # Read the pickle file
    data = torch.load(dataset_path, weights_only=False)
    
    # Print dataset information
    if isinstance(data, Data):
        print("Dataset Information:")
        print(f"Number of nodes: {data.num_nodes}")
        print(f"Number of edges: {data.num_edges}")
        print(f"Number of node features: {data.num_node_features}")
        print(f"Number of classes: {len(torch.unique(data.y))}")
        print(f"\nAvailable attributes:", data.keys)
        print(f"\nNode features shape:", data.x.shape)
        print(f"Edge index shape:", data.edge_index.shape)
        print(f"Labels shape:", data.y.shape)
    else:
        print("Data format:", type(data))
        print("Content:", data.keys())
        print(data['tr_omic'].shape)
        print(data['data_tr'].shape)

if __name__ == "__main__":
    read_ba_shapes("ROSMAP/data.pt")
```

```python
import torch
from torch_geometric.data import Data

def analyze_masks(dataset_path):
    data_tuple = torch.load(dataset_path, weights_only=False)
    data = data_tuple[0]
    
    print("Mask Analysis:")
    print(f"Total nodes: {data.num_nodes}")
    print(f"\nTraining nodes: {data.train_mask.sum().item()}")
    print(f"Validation nodes: {data.val_mask.sum().item()}")
    print(f"Test nodes: {data.test_mask.sum().item()}")
    
    # Check overlap
    print("\nMask Overlap Check:")
    train_and_val = (data.train_mask & data.val_mask).sum().item()
    train_and_test = (data.train_mask & data.test_mask).sum().item()
    val_and_test = (data.val_mask & data.test_mask).sum().item()
    
    print(f"Nodes in both train and val sets: {train_and_val}")
    print(f"Nodes in both train and test sets: {train_and_test}")
    print(f"Nodes in both val and test sets: {val_and_test}")

if __name__ == "__main__":
    analyze_masks("ROSMAP/data.pt")
```
